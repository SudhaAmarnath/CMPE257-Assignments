```
import requests
import re
re.compile('<title>(.*)</title>')
page = requests.get("https://www.theverge.com/2019/10/3/20897560/vaping-lung-injury-toll-cdc-investigation-fda-deaths")
page.content
```

# Parse the web content and scrape the data out of HTML file

```
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

## Get paragraphs associated to the content relating age and % affected categories

```
paragraph = soup.find('p', id='h52VJI').get_text()
print(paragraph)
```

```
lines = re.split(r'[`\-=~!@#$%^&*()_+\[\]{};\'\\:"|<,./<>?]', paragraph)
print(lines)
```

```
percent_list = []
yeartype_list = []
import pandas as pd
for sentences in lines:
  
    percent = re.search(r'[0-9]{0,2} percent', sentences, flags=re.I)
    year_type1 = re.search(r'[0-9]{0,5} to [0-9]{0,5} years', sentences, flags=re.I)
    year_type2 = re.search(r'under [0-9]{0,4} years', sentences, flags=re.I)
    year_type3 = re.search(r'[0-9]{0,4} years or older', sentences, flags=re.I)
    if percent:
        percent_list.append(percent.group())
        if year_type1:
            yeartype_list.append(year_type1.group())
        elif year_type2:
            yeartype_list.append(year_type2.group())
        elif year_type3:
            yeartype_list.append(year_type3.group())
        else :
            yeartype_list.append("NA")
```

```
df = pd.DataFrame(list(zip(percent_list, yeartype_list)), 
               columns =['Percent', 'Years']) 
df
```

## The age category between 25-34 appears to be the most affected by Vaping.
